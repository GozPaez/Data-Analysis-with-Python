## Use PLOT and SCATTER

```python
import matplotlib.pyplot as plt
```

```python
year = [1960, 1970, 1980, 1990, 2000, 2010, 2020, 2030, 2040]
```

```python
pop = [2.516, 10.692, 5.263, 6.972, 5.689, 2.589, 13.562, 5.569, 7.597]
```

```python
plt.plot(year, pop)
```

```python
plt.scatter(year, pop)
```

```python
print ("last item of year list: " , year[-1], "last item of pop list: " , pop[-1])
```

When you are trying to assess if there's a correlation between two variables, for example, the scatter plot is the better choice.
